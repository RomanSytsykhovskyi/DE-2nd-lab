```python
def csv(filename, unusedColumns, amountColumnts):
    cols = [i for i in range(amountColumnts) if i not in unusedColumns]
    csvf = pd.read_csv(os.path.join(__dirname, filename), usecols=cols)

    return csvf
```

Data cleaning, also called data cleansing or scrubbing, deals with detecting and removing errors and
inconsistencies from data in order to improve the quality of data. Data quality problems are present in single
data collections, such as files and databases, e.g., due to misspellings during data entry, missing information
or other invalid data. When multiple data sources need to be integrated, e.g., in data warehouses, federated
database systems or global web-based information systems, the need for data cleaning increases
significantly. This is because the sources often contain redundant data in different representations. In order to
provide access to accurate and consistent data, consolidation of different data representations and elimination
of duplicate information become necessary.

```python
def randomLetters():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def generate():
    brand = random.randint(0, len(brands))
    brand = randomLetters() if brand == len(brands) else brands[brand]
    
    model = random.randint(0, len(models))
    model = randomLetters() if model == len(models) else models[model]
    
    serie = random.randint(0, len(series))
    serie = randomLetters() if serie == len(series) else series[serie]
    
    processor = random.randint(0, len(processors))
    processor = randomLetters() if processor == len(processors) else processors[processor]
    
    processor_gen = random.randint(0, len(processor_gens))
    processor_gen = randomLetters() if processor_gen == len(processor_gens) else processor_gens[processor_gen]
    
    ram = random.randint(0, len(rams))
    ram = randomLetters() if ram == len(rams) else rams[ram]
    
    hardDiskCapacity = random.randint(0, len(hardDisksCapacity))
    hardDiskCapacity = randomLetters() if hardDiskCapacity == len(hardDisksCapacity) else hardDisksCapacity[hardDiskCapacity]
    
    OS = random.randint(0, len(OSs))
    OS = randomLetters() if OS == len(OSs) else OSs[OS]
    
    rating = random.randint(0, 4) + random.randint(0, 10) / 10
    
    stdPrice = numpy.std(csvDataframe["Price"])
    price = numpy.mean(csvDataframe["Price"]) + random.triangular(-stdPrice, stdPrice)
    
    
    return [brand, model, serie, processor, processor_gen, ram, hardDiskCapacity, OS, rating, price]

generate()
```

The way generating of data and mistakes might be different, like entry datas.

```python
entryArray = []
for i in range(100):
    entryArray.append(generate())
    
entryArray
```

```python
entryArray = pd.DataFrame(entryArray, columns=columnsName)

csvDataframe = csvDataframe.append(entryArray)

csvDataframe = csvDataframe.reset_index(drop=True)
```

```python
csvDataframe.loc[~csvDataframe["Brand"].isin(models), "Brand"] = "Missing"

for brand in brands:
    csvDataframe.drop(csvDataframe[(csvDataframe["Brand"] == brand) & (~csvDataframe["Model"].isin(brandModels[brand]))].index, inplace=True)    
csvDataframe.loc[~csvDataframe["Model"].isin(models), "Model"] = "Missing"
csvDataframe.loc[~csvDataframe["Series"].isin(series), "Series"] = "Missing"
csvDataframe.loc[~csvDataframe["Processor"].isin(processors), "Processor"] = "Missing"
csvDataframe.loc[~csvDataframe["Processor_Gen"].isin(processor_gens), "Processor_Gen"] = "Missing"
csvDataframe.loc[~csvDataframe["RAM"].isin(rams), "RAM"] = "Missing"
csvDataframe.loc[~csvDataframe["Hard_Disk_Capacity"].isin(hardDisksCapacity), "Hard_Disk_Capacity"] = "Missing"
csvDataframe.loc[~csvDataframe["OS"].str.contains("|".join(OSs), case=False, na=False), "OS"] = "Missing"

csvDataframe = csvDataframe.drop_duplicates()

csvDataframe
```

```python
csvDataframe.to_csv(r'result.csv', index=False, sep=',')
```
