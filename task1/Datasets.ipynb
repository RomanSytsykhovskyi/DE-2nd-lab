```python
import pandas as pd
import os
import time
import numpy
##import scipy.stats as st
import math
```

```python
__dirname = os.getcwd()
```

```python
csvName = 'Cleaned_Laptop_data.csv'
jsonName = "Cleaned_Laptop_data.json"
```

```python
def csv(filename):
    csvf = pd.read_csv(os.path.join(__dirname, filename))

    return csvf
```

```python
%%time
fileCsv = csv(csvName)

fileCsv
```

I used csd because it's a familiar format for me

```python
def json(filename):
    jsonf = pd.read_json(os.path.join(__dirname, filename))

    return jsonf
```

```python
%%time
fileJson = json(jsonName)

fileJson
```

I used JSON because I often works with it. It is my favorite format because it's a very popural one for api requests like xml.

```python
def timeArray(func, filename, quantity):
    delay = []
    for i in range(quantity):
        startTime = time.time() 
        func(filename)
        delta = time.time() - startTime
        delay.append(delta)
    return delay
        
```

```python
def toFixed(numObj, digits=6):
    return f"{numObj:.{digits}f}"
```

```python
csvTimeArray = timeArray(csv, csvName, 100)

csvTimeArray
```

```python
averageCsv = numpy.mean(csvTimeArray)

deviation = numpy.std(csvTimeArray)

print(f"Execution time of reading that {csvName} file is {toFixed(averageCsv)}±{toFixed(deviation)}s")
```

```python
jsonTimeArray = timeArray(json, jsonName, 100)

jsonTimeArray
```

```python
averageJson = numpy.mean(jsonTimeArray)

deviation = numpy.std(jsonTimeArray)

print(f"Execution time of reading that {jsonName} file is {toFixed(averageJson)}±{toFixed(deviation)}s")
```

```python
if averageCsv < averageJson:
    print(f"Average time of reading that {jsonName} file is higher by {100 * float(toFixed((averageJson - averageCsv) / averageCsv))}% than the csv one")
else:
    print(f"Average time of reading that {csvName} file is higher by {100 * float(toFixed((averageCsv - averageJson) / averageJson))}% than the json one")
```

```python
csvSize = os.path.getsize(os.path.join(__dirname, csvName))

print(f"The {csvName} file size is {csvSize}b")
```

```python
jsonSize = os.path.getsize(os.path.join(__dirname, jsonName))

print(f"The {jsonName} file size is {jsonSize}b")
```

```python
if csvSize < jsonSize:
    print(f"The size of that {jsonName} file is bigger by {100 * float(toFixed((jsonSize - csvSize) / csvSize))}% than the csv one")
else:
    print(f"The size of that {csvName} file is bigger by {100 * float(toFixed((csvSize - jsonSize) / jsonSize))}% than the csv one")
```

```python
csvMedian = numpy.median(csvTimeArray)

csvP99 = sorted(csvTimeArray)[-math.floor(len(csvTimeArray) / 100):]

print(f"The median is {csvMedian} and 99 percentile is {csvP99} for {csvName}")
```

```python
jsonMedian = numpy.median(jsonTimeArray)

jsonP99 = sorted(jsonTimeArray)[-math.floor(len(jsonTimeArray) / 100):]

print(f"The median is {jsonMedian} and 99 percentile is {jsonP99} for {jsonName}")
```

The both json and csv are a text format (csv is a file format from other sources) but the main feature of json format is that it is nested what causes its bigger size than the csv one. The higher time of execution the json file is caused by its size. Although, the speed of reading json file is higher (27,788,765.9b/s against 22,756,728.77).

P.S.: the speed was calculated within the one experiment over one dataset. The way of speed calculating is dividing the size of file by the time of reading.
